```python
# Merge the dataframes
pyber_data_df = pd.merge(ride_data_df, city_data_df, on=["city", "city"], how="left")

# Display the head
pyber_data_df.head()
```

```python
# Create individual dataframes for each city type

urban_df = pyber_data_df[pyber_data_df["type"]=="Urban"]
rural_df = pyber_data_df[pyber_data_df["type"]=="Rural"]
suburban_df = pyber_data_df[pyber_data_df["type"]=="Suburban"]
```

```python
# Group by the city name and get the count of rides

urban_ride_count = urban_df.groupby(['city']).count()['ride_id']
rural_ride_count = rural_df.groupby(['city']).count()['ride_id']
suburban_ride_count = suburban_df.groupby(['city']).count()['ride_id']
```

```python
# Group by the city name and get the mean fare by city

urban_avg_fare = urban_df.groupby(['city']).mean()['fare']
rural_avg_fare = rural_df.groupby(['city']).mean()['fare']
suburban_avg_fare = suburban_df.groupby(['city']).mean()['fare']
```

```python
# Group by the city name and get the mean driver count

urban_driver_count = urban_df.groupby(['city']).mean()['driver_count']
rural_driver_count = rural_df.groupby(['city']).mean()['driver_count']
suburban_driver_count = suburban_df.groupby(['city']).mean()['driver_count']
```

```python
urban_driver_count.head()
```

# Creating Bubble Chart

```python
# Create bubble chart (Urban)
plt.scatter(urban_ride_count, urban_avg_fare, s=urban_driver_count*10,
            color='coral', edgecolors="black", linewidth=1, label="Urban")
plt.title("Urban City Rides (2019)")
plt.xlabel("Ride Count")
plt.ylabel("Average Fare ($)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Create bubble chart (Suburban)
plt.scatter(suburban_ride_count, suburban_avg_fare, s=suburban_driver_count*10,
            color='skyblue', edgecolors="black", linewidth=1, label="Suburban")
plt.title("Suburban City Rides (2019)")
plt.xlabel("Ride Count")
plt.ylabel("Average Fare ($)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Create bubble chart (Rural)
plt.scatter(rural_ride_count, rural_avg_fare, s=rural_driver_count*10,
            color='gold', edgecolors="black", linewidth=1, label="Rural")
plt.title("Rural City Rides (2019)")
plt.xlabel("Ride Count")
plt.ylabel("Average Fare ($)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Add the scatter charts for each type of city.
plt.subplots(figsize=(10, 6))
plt.scatter(urban_ride_count,
      urban_avg_fare,
      s=10*urban_driver_count, c="coral",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Urban")

plt.scatter(suburban_ride_count,
      suburban_avg_fare,
      s=10*suburban_driver_count, c="skyblue",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Suburban")

plt.scatter(rural_ride_count,
      rural_avg_fare,
      s=10*rural_driver_count, c="gold",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Rural")

# Add title / labels / legend / resize
plt.title("PyBer Ride Share Data by City (2019)", fontsize=20)
plt.xlabel("Driver Count", fontsize=12)
plt.ylabel("Average Fare ($)", fontsize=12)
plt.grid(True)

# Create a legend
lgnd = plt.legend(fontsize="12", mode="Expanded",
         scatterpoints=1, loc="best", title="City Types")
lgnd.legendHandles[0]._sizes = [75]
lgnd.legendHandles[1]._sizes = [75]
lgnd.legendHandles[2]._sizes = [75]
lgnd.get_title().set_fontsize(12)

# Incorporate a text label about circle size.
plt.text(42, 35, "Note: Circle size correlates with driver count per city.", fontsize="12")

# Save the figure.
plt.savefig("analysis/Fig1.png")

# Show the plot
plt.show()
```

# Section 5.5 -- Percentage of Fares Pie

```python
# Group pyber_data_df by city and sum revenue

sum_fares_by_type = pyber_data_df.groupby(["type"]).sum()["fare"]
sum_fares_by_type
```

```python
# Get the sum of all the fares.
total_fares = pyber_data_df["fare"].sum()
total_fares
```

```python
# Calculate the percentage of fare for each city type.
type_percents = 100 * sum_fares_by_type / total_fares
type_percents
```

```python
# Import mpl to change the plot configurations using rcParams.
import matplotlib as mpl

# Build the percentage of fares by city type pie chart.
plt.subplots(figsize=(10, 6))
plt.pie(type_percents,
    labels=["Rural", "Suburban", "Urban"],
    colors=["gold", "lightskyblue", "lightcoral"],
    explode=[0, 0, 0.1],
    autopct='%1.1f%%',
    shadow=True, startangle=150)
plt.title("% of Total Fares by City Type")

# Change the default font size from 10 to 14.
mpl.rcParams['font.size'] = 14

# Save Figure
plt.savefig("analysis/Fig5.png")

# Show Figure
plt.show()
```
